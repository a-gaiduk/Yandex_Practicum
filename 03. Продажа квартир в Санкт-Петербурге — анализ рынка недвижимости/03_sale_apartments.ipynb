```python
# Построим в цикле графики по всем числовым переменным
for column_name, column_type in data.dtypes.iteritems():
    if column_type in ('int64', 'float64'):
        data.plot(
            kind='hist', 
            y=column_name, 
            bins=50, 
            figsize=(12,5), 
            title='Гистограмма по столбцу: ' + columns_dict.get(column_name),
            fontsize = 12
        )
        plt.xlabel(columns_dict.get(column_name), fontsize = 12)
        plt.ylabel('Количество', fontsize = 12)
        plt.show();
    if 'datetime' in str(column_type):
        data[column_name].hist(
            bins=50, 
            figsize=(12,5)
        )
        plt.xlabel(columns_dict.get(column_name), fontsize = 12)
        plt.ylabel('Количество', fontsize = 12)
        plt.gca().set_title('Гистограмма по столбцу: ' + columns_dict.get(column_name))
        #plt.rc('font', size=10)
        plt.show();
```

```python
data_clean = data.copy()

print('Всего ответов в столбце living_area до восстановления: {}'.format(data_clean['living_area'].count()))
data_clean['living_area'] = data_clean['living_area'].where(data_clean['living_area']<data_clean['total_area'], data_clean['total_area'])
print('Всего ответов в столбце living_area после восстановления: {}\n'.format(data_clean['living_area'].count()))

print('Всего ответов в столбце balcony до восстановления: {}'.format(data_clean['balcony'].count()))
data_clean['balcony'] = data_clean['balcony'].fillna(0)
print('Всего ответов в столбце balcony после восстановления: {}\n'.format(data_clean['balcony'].count()))

print('Всего ответов в столбце parks_around3000 до восстановления: {}'.format(data_clean['parks_around3000'].count()))
data_clean['parks_around3000'] = data_clean['parks_around3000'].fillna(0)
print('Всего ответов в столбце parks_around3000 после восстановления: {}'.format(data_clean['parks_around3000'].count()))
print('Количество тех, кто указал расстояние до ближайшего парка меньше 3 км, а в количестве парков до 3 км указали 0: {}\n'
      .format(data_clean.query('parks_nearest<3000 and parks_around3000==0')['parks_nearest'].count()))

print('Всего ответов в столбце ponds_around3000 до восстановления: {}'.format(data_clean['ponds_around3000'].count()))
data_clean['ponds_around3000'] = data_clean['ponds_around3000'].fillna(0)
print('Всего ответов в столбце ponds_around3000 после восстановления: {}'.format(data_clean['ponds_around3000'].count()))
print('Количество тех, кто указал расстояние до ближайшего водоема меньше 3 км, а в количестве водоемов до 3 км указали 0: {}\n'
      .format(data_clean.query('ponds_nearest<3000 and ponds_around3000==0')['ponds_nearest'].count()))

print('Всего ответов в столбце is_apartment до восстановления: {}'.format(data_clean['is_apartment'].count()))
data_clean['is_apartment'] = data_clean['is_apartment'].fillna(False)
print('Всего ответов в столбце is_apartment после восстановления: {}\n'.format(data_clean['is_apartment'].count()))
```

<div style="border:solid gray 0.2px; padding: 10px"> 

**Построим графики**, чтобы определить факторы, от которых зависит стоимость квартиры
    
    
Т.к. в данных встречаются нетипичные значения, минимум и максимум для диаграмм рассеяния будем определять с помощью фннкции с параметрами:
    
- **Максимум** как среднее + n стандартных отклонений
    
- **Минимум** как максимум из: среднего - n стандартных отклонения и нуля

<div>

```python
def get_lim(columns, num_std, isx = True, isy=True):

    ylim_max = data_rec['last_price'].mean() + data_rec['last_price'].std() *num_std
    ylim_min = max(0, data_rec['last_price'].mean() - data_rec['last_price'].std() *num_std)
    xlim_max = data_rec[columns].mean() + data_rec[columns].std() *num_std
    xlim_min = max(0, data_rec[columns].mean() - data_rec[columns].std() *num_std)
    
    return ylim_max, ylim_min, xlim_max, xlim_min
```

    - в городе Пушник квартиры по стоимости сравнимы с Санкт-Петербургом, а в Гатчине самые доступные квартиры из ТОП10 населенных пунктов
